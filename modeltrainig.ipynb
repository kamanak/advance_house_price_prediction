```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.pandas.set_option('display.max_columns',None)
```

```python
# Read feature engineering train data 
dataset = pd.read_csv('X_train.csv')
```

```python
dataset
```

```python
# splitting data into dependent and independent feature 
y_train = dataset[['SalePrice']]
X_train = dataset.drop(['Id','SalePrice'],axis=1)
```

Feature Selection

```python

from sklearn.linear_model import LinearRegression


selecter = SelectFromModel(Lasso(alpha=0.001))
selecter.fit(X_train,y_train)
```

```python
selecter.get_support()
```

```python
selecter.estimator_.coef_
```

```python
selected_feat = X_train.columns[(selecter.get_support())]
X= X_train[selected_feat]
X
```

84 different features are reducted into 12 features using lasso regression

```python

y = y_train
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
model = LinearRegression()
model = model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print('Mean Squared error: ',mean_squared_error(y_test,y_pred))
print('R2 Score: ', r2_score(y_test,y_pred)) 
```

Random forest Regressior

```python
value = pd.read_csv('X_train.csv')
value
```

```python
X = value.drop(['Id','SalePrice'],axis=1)
y = value.SalePrice
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV

rf = RandomForestRegressor()
```

```python
# Number of trees in randomForest

n_estimaters = [int(x) for x in np.linspace(start=200,stop=2000,num=10)]
n_estimaters

# Number of features consider at every split
max_features = ['auto','sqrt','log2']

# Maximum numers of level in the tree
max_depth = [int(x) for x in np.linspace(10,1000,10)]

#Minimum numbers of samples required to split the node
min_samples_split = [1,2,3,4,6,7,8]

# Minimun nubers of samples required at each leaf node
min_samples_leaf =[1,2,4,6,8]

random_grid = {
    'n_estimators':n_estimaters,
    'max_features':max_features,
    'max_depth':max_depth,
    'min_samples_split':min_samples_split,
    'min_samples_leaf':min_samples_leaf,
}
print(random_grid)
```

```python
rf_randomCV = RandomizedSearchCV(rf,param_distributions=random_grid,n_iter=100,cv=3,verbose=2,random_state=100,n_jobs=-1)
rf_randomCV.fit(X_train,y_train)
```

```python
rf_randomCV.best_params_
```

```python
rf_randomCV.best_estimator_
```

```python
best_random_grid = rf_randomCV.best_estimator_
```

```python
from sklearn import metrics
y_predict = best_random_grid.predict(X_test)
print('Mean Squared error: ',mean_squared_error(y_test,y_predict))
print('R2 Score: ', r2_score(y_test,y_predict))
# cm = metrics.confusion_matrix(y_test,y_predict)
# print(cm)
# accuracy = metrics.accuracy_score(y_test,y_predict)
# print(accuracy)
# report = metrics.classification_report(y_test,y_predict)
# print(report)
```
